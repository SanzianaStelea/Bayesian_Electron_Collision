### Better looking plots

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
import scipy
import scipy.optimize as opt
```

```python
# Get the data from the file
def get_data(filename):
    data = pd.read_csv(filename, sep=',')
    M_data = data['M'].values

    y, bins, _ = plt.hist(M_data, bins=200, range=(4,125), alpha=0.5, label='Histogram of the data')
    plt.close()
    x = 0.5*(bins[1:]+bins[:-1])
    y_errors = np.sqrt(y)  # Assuming Poisson statistics for the errors ----MAYBE ASK ABOUT THIS----

    return x, y, y_errors
```

```python
# Plot the data
def plot_data(x, y, y_errors):
    plt.errorbar(x, y, yerr=y_errors, fmt='.',color = 'b', label='Data')
    plt.plot(x, y, linewidth=0.5, color = 'b')
    plt.xlabel('Invariant mass [GeV]')
    plt.ylabel('Counts')
    plt.grid()
    plt.legend()
    plt.show()
```

```python
def plot_prior_model_predictive(x, y, prior_model_predictive):
    style = dict(c="C2", lw=1, alpha=0.5)
    plt.plot(x, y, '.', color = 'black')
    plt.plot(x, prior_model_predictive.T, **style)
    plt.plot([], [], label="Prior model predictive", **style)
    plt.legend()
```
